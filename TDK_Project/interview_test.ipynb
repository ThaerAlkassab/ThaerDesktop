```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Example usage:
num1 = 36
num2 = 48
print("GCD of", num1, "and", num2, "is:", gcd(num1, num2))
```

```python
def lcm(a, b):
    return abs(a*b) // gcd(a, b)

# Example usage:
num1 = 36
num2 = 48
print("LCM of", num1, "and", num2, "is:", lcm(num1, num2))
```
